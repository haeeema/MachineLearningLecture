### Logistic Regression - '회귀지만 하는일은 분류'

```python
import numpy as np
import matplotlib.pyplot as plt
```

- Sigmoid function

```python
def sigmoid(x, a=1, b=0):
    return 1 / (1 + np.exp(-a * (x - b)))
    # return 1 / (1 + np.exp(-(a*x + b)))
```

```python
xs = np.linspace(-5, 5, 1001)
ys = sigmoid(xs)
plt.figure(figsize=(5, 3))
plt.plot(xs, ys, label='sigmoid')
plt.plot(xs, ys*(1-ys), label='derivative')
plt.title('Sigmoid function')
plt.yticks([0, 0.5, 1])
plt.grid(), plt.legend()
plt.show()
```

```python
# a값에 따른 sigmoid 함수의 변화
xs = np.linspace(-5, 5, 1001)
y3 = sigmoid(xs, a=3)
y_half = sigmoid(xs, a=0.5)
y_99 = sigmoid(xs, a=99)
plt.figure(figsize=(5, 3))

plt.plot(xs, ys, label='a=1')
plt.plot(xs, y3, label='a=3')
plt.plot(xs, y_half, label='a=0.5')
plt.plot(xs, y_99, label='a=99')

plt.title('Sigmoid function')
plt.yticks([0, 0.5, 1])
plt.grid(), plt.legend()
plt.show()
```

```python
# b값에 따른 sigmoid 함수의 변화
xs = np.linspace(-5, 5, 1001)
y_plus2 = sigmoid(xs, b=2)
y_minus2 = sigmoid(xs, b=-2)
plt.figure(figsize=(5, 3))

plt.plot(xs, ys, label='b=0')
plt.plot(xs, y_plus2, label='b=2')
plt.plot(xs, y_minus2, label='b=-2')

plt.title('Sigmoid function')
plt.yticks([0, 0.5, 1])
plt.grid(), plt.legend()
plt.show()
```

### 이진 분류

```python
import pandas as pd
pima = pd.read_csv('../01.소개/pima-indians-diabetes.csv', skiprows=9, header=None)
pima.columns = ['P', 'G', 'BP', 'S', 'I', 'BMI', 'D', 'Age', 'Target']
pima.head()
```

```python
X = pima.iloc[:, :-1].values
y = pima.Target.values
# y = pima.iloc[:, -1].values
```

```python
# 로지스틱 회귀를 사용하려면 데이터를 표준화/정규화 해야 한다.
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X)
```

```python
# 데이터셋 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_std, y, stratify=y, test_size=0.2, random_state=2023
)
```

```python
# 로지스틱 회귀 하이퍼 파라메터
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression(random_state=2023)
lrc.get_params()
```

```python
lrc.fit(X_train, y_train)
lrc.score(X_test, y_test)
```

```python
# 선형회귀의 가중치(계수)
lrc.coef_       # coefficient: 가중치, 계수
```

```python
# 선형회귀의 바이어스(절편)
lrc.intercept_      # intercept: 절편
```

```python
X_test[0], y_test[0]
```

```python
# X_test[0]가 1이 될 확률, sigmoid 함수 적용 전 값
val = np.dot(lrc.coef_, X_test[0]) + lrc.intercept_
val
```

```python
# X_test[0]가 1이 될 확률, sigmoid 함수 적용 후 값
sigmoid(val)
```

```python
# lrc.predict_proba(X_test[0].reshape(1, -1))
lrc.predict_proba(X_test[:1])
```

### 다중분류

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris_std = StandardScaler().fit_transform(iris.data)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_std, iris.target, stratify=iris.target, test_size=0.2, random_state=2023
)
```

```python
lrc = LogisticRegression(random_state=2023)
lrc.fit(X_train, y_train)
lrc.score(X_test, y_test)
```

```python
lrc.predict_proba(X_test[:1]) 
```

```python
# weight, bias
lrc.coef_, lrc.intercept_
```

```python
# X_test[0]가 1이 될 확률, sigmoid 함수 적용 전 값
val = np.dot(lrc.coef_, X_test[0]) + lrc.intercept_
val
```

```python
# sigmoid 함수 적용 후 값
sigmoid(val)
```

```python
# softmax 함수
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python
softmax(np.array([1, 2, 3]))
```

```python
# 다중 분류에서는 val에 softmax를 적용한 값이 predict_proba와 같다.
softmax(val)
```
